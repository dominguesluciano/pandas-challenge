```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
hop_file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(hop_file)
purchase_data.head()
```

```python
hop_df = pd.DataFrame(purchase_data)
hop_df = hop_df.drop_duplicates()
```

## Player Count

* Display the total number of players


```python
#count Unique values of column SN (Player Name)
sn_count = purchase_data["SN"].nunique()

#create dataframe
total_num_players = pd.DataFrame([{"Total Number of Players":sn_count}])
total_num_players

```

```python
hop_df.columns
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#count of unique Items
ii_count = hop_df["Item ID"].nunique()

#average price
avg_price = hop_df["Price"].mean()

#count of purchases
count_purchases = hop_df["Purchase ID"].count()

#sum of purchases (revenue)
sum_purchases = hop_df["Price"].sum()


summary_df = [ii_count,avg_price,count_purchases,sum_purchases]
p_analysis = pd.DataFrame([summary_df], columns = ['Number of Unique Items','Average Price','Number of Purchases','Total Revenue']) 
p_analysis.round(2)
p_analysis['Average Price'] = p_analysis['Average Price'].map("${:,.2f}".format)
p_analysis['Total Revenue'] = p_analysis['Total Revenue'].map("${:,.2f}".format)
p_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_demographics = hop_df.groupby(['Gender'])
total_count_sn = gender_demographics["SN"].nunique()
pct_sn = (total_count_sn/sn_count)*100
summary_df = pd.DataFrame({"Total Count" : total_count_sn,
                           "Percentage of Players" : pct_sn})
summary_df["Percentage of Players"] = summary_df["Percentage of Players"].map("{:.2f}%".format)
summary_df = summary_df.sort_values(by="Total Count", ascending = False)
summary_df
summary_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_grouped_mean = hop_df.groupby(['Gender']).mean()['Price']
gender_grouped_count = hop_df.groupby(['Gender']).count()['Price']
gender_grouped_sum = hop_df.groupby(['Gender']).sum()['Price']
avg_per_person = gender_grouped_sum/total_count_sn


summary_df = pd.DataFrame({"Purchase Count":gender_grouped_count,"Average Purchase Price":gender_grouped_mean,
                                "Total Purchase Value":gender_grouped_sum,"Avg Total Purchase per Person":avg_per_person})
summary_df["Average Purchase Price"] = summary_df["Average Purchase Price"].map("${:.2f}".format)
summary_df["Total Purchase Value"] = summary_df["Total Purchase Value"].map("${:.2f}".format)
summary_df["Avg Total Purchase per Person"] = summary_df["Avg Total Purchase per Person"].map("${:.2f}".format)

summary_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

hop_df["Age Brackets"] = pd.cut(hop_df["Age"], age_bins, labels=age_ranges)
age_demo = hop_df.groupby("Age Brackets")

age_brackets_count = age_demo["SN"].nunique()
pct_age_players = age_brackets_count/sn_count*100

summary_df = pd.DataFrame({"Total Count":age_brackets_count,"Percentage of Players":pct_age_players})
#summary_df["Percentage of Players"] = summary_df["Percentage of Players"].map("{:.2f}%".format)

summary_df.round(2).sort_index()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spender = hop_df.groupby("SN")
purchase_count = top_spender["Purchase ID"].count()
avg_purchase_price = top_spender["Price"].mean()
total_purchase_value = top_spender["Price"].sum()

summary_df = pd.DataFrame({"Total Count":purchase_count,
                           "Average Purchase Price":avg_purchase_price,
                           "Total Purchase Values":total_purchase_value
                           })

summary_df = summary_df.sort_values(by="Total Purchase Values", ascending = False)

summary_df["Average Purchase Price"] = summary_df["Average Purchase Price"].map("${:.2f}".format)
summary_df["Total Purchase Values"] = summary_df["Total Purchase Values"].map("${:.2f}".format)

summary_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
mpopi = hop_df.groupby(["Item ID","Item Name"])
purchase_count = mpopi["Purchase ID"].count()
avg_item_price = mpopi["Price"].mean() # had to aggreate to get item price... 
total_purchase_value = mpopi["Price"].sum()
summary_df = pd.DataFrame({"Total Count":purchase_count,
                           "Item Price":avg_item_price,
                           "Total Purchase Values":total_purchase_value
                           })
summary_df = summary_df.sort_values(by="Total Count", ascending = False)
summary_df["Item Price"] = summary_df["Item Price"].map("${:.2f}".format)
summary_df["Total Purchase Values"] = summary_df["Total Purchase Values"].map("${:.2f}".format)

summary_df.head()
```
